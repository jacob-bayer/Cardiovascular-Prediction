```python
## gender is binary but the values is 1 and 2, swap 2 with 0 for starndard binary representation
cardio_data.gender.replace(to_replace=2, value =0, inplace=True)
```

```python
## scatterplot of height and weight to visualize potential out of range values.
sns.scatterplot(x= cardio_data.height, y = cardio_data.weight, hue = cardio_data.gender, alpha =0.8)
plt.title("scatterplot of weight vs height")
```

### First exam the height. there are small clusters of short people. But there are people with medical conditions which resulting in short statues. By definition, dwarfism is considered for people less than 147cm.[1] I think we should not exclude these data because the heights are uncommon. Also, there is one data which is very tall at 250cm, it seems possible because there are several record of human with similar heights. However, this record is female, there is no female ever in record with this height.[2] We should exclude this datapoint.

```python
## drop the data with height 250cm
cardio_data = cardio_data[cardio_data.height!=250].copy()
```

```python
cardio_data.shape
```

### There are several observations with abnormal weight/height ratios. For example, several data on the upper left corner indicates short statues and high weight, and several data with extremly low weights. We could looks at this data when we create a weight/height ratio. BMI is a common metric to measure weight/height ratio.

```python
## create bmi feature
cardio_data["bmi"] = cardio_data.weight / (cardio_data.height/100)**2
```

```python
## distribution of bmi
cardio_data.bmi.hist()
plt.title("distribution of bmi")
plt.show()
```

```python
## there are several extreme bmi
cardio_data.bmi.describe()
```

```python
sns.boxplot(x = cardio_data.gender, y = cardio_data.bmi)
plt.title("boxplot of bmi")
plt.show()
```

### The bmi has several unrealist outliers. These outliers should be excluded.

```python
## exclude outlier out of 1.5*IQR
IQR = cardio_data.bmi.quantile(0.75)-cardio_data.bmi.quantile(0.25)
upper = cardio_data.bmi.quantile(0.75) + 1.5 * IQR
lower = cardio_data.bmi <= cardio_data.bmi.quantile(0.25) - 1.5 * IQR

cardio_1 = cardio_data[(cardio_data.bmi > lower) & (cardio_data.bmi < upper)]

cardio_1.shape
```

```python
cardio_1.shape[0]/cardio_data.shape[0]
```

### Exclude about 3% data based on abnormal bmi

### ap_hi and ap_lo are measurement about blood pressure. It is impossible to be negative

```python
## scatterplot of systolic blood pressure and diastolic blood pressure .
sns.scatterplot(x= cardio_1.ap_lo, y = cardio_1.ap_hi, hue = cardio_1.gender, alpha =0.8)
plt.title("scatterplot of Diastolic blood pressure vs Systolic blood pressure")
plt.ylabel("Systolic blood pressure")
plt.xlabel('Diastolic blood pressure')
```

### The data about blood pressure is messy. 
### 1.Some medical professional suggest systolic blood pressure higher than 180 or diastolic blood pressure greater than 100 requires immediate medical attention.The data source is ambiguous about where the data was collected, I can't rule out the possibility of the data is from hospital. Giraffe has the  highest systolic blood pressure with 300 among all species of mammal. We will use it for the upper limit of systolic blood pressure. [3]
### 2, it is impossible to have negative blood pressure.
### 3. Systolic blood pressure always higher than diastolic blood pressure.

```python
## exclude unrealistic data
cardio_2 = cardio_1[(cardio_1.ap_hi < 300) & (cardio_1.ap_hi > 0) & (cardio_1.ap_lo < cardio_1.ap_hi) & (cardio_1.ap_lo > 0)]
cardio_2.shape
```

```python
## scatterplot of systolic blood pressure and diastolic blood pressure .
sns.scatterplot(x= cardio_2.ap_lo, y = cardio_2.ap_hi, hue = cardio_2.gender, alpha =0.8)
plt.title("scatterplot of Diastolic blood pressure vs Systolic blood pressure")
plt.ylabel("Systolic blood pressure")
plt.xlabel('Diastolic blood pressure')
```

```python
## boxplot of the systolic and diastolic blood pressure

sns.boxplot(data =cardio_2.loc[:, ['ap_lo', 'ap_hi']].melt(), x='variable', y='value')
```
